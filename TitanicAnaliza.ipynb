```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #do wykresow
import seaborn as sns #do jeszcze ladniejszych wykresow

r_filenameTSV = './TitanicCleaned.tsv'

print(r_filenameTSV)


# Zaczytanie danych
db = pd.read_csv(r_filenameTSV, sep='\t')

# Usunięcie duplikatów
print(pd.crosstab(db["Survived"],db.Survived))
sns.set_palette("Set1")
sns.catplot(data=db, x='Survived', kind='count')
plt.show()

print(pd.crosstab(db["Pclass"],db.Survived))
ax = sns.relplot(x = 'Pclass', y = 'Survived', kind="line" ,palette = 'Set1', data = db)
ax.set(title = 'Przeżywalność w stosunku do klasy', 
       xlabel = 'Klasa pasażera', ylabel = 'Liczba')
plt.show()

print(pd.crosstab(db["Age"],db.Survived))
ax = sns.relplot(x = 'Age', y = 'Survived', palette = 'Set1', data = db)
ax.set(title = 'Przeżywalność w stosunku do wieku', xlabel = 'Płeć', ylabel='Liczba')
plt.show()

print(pd.crosstab(db['Embarked'], db.Survived))
ax = sns.countplot(x = 'Embarked', hue = 'Survived', palette = 'Set1', data = db)
ax.set(title = 'Przeżywalność w stosunku do miejsca wsiadania na pokład')
plt.show()

print(pd.crosstab(db['Sex'], db.Survived))
ax = sns.countplot(x = 'Sex', hue = 'Survived', palette = 'Set1', data = db)
ax.set(title = 'Przeżywalność w stosunku do płci')
plt.show()
```
